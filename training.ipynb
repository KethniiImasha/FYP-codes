```python
# make all image inside every folder in this folder grayscale and resize to 256x256 then copy to a new folder
import cv2
import os
import glob
from pathlib import Path
from tqdm import tqdm

dataset_path =r"/Users/kethniimasha/Desktop/FYP Final Project/2000_per_class_256x256_Grayscale"
output_path = r"/Users/kethniimasha/Desktop/FYP Final Project/2000_per_class_256x256_Grayscale_split"

for class_folder in os.listdir(dataset_path):
    class_path = os.path.join(dataset_path, class_folder)
    output_class_path = os.path.join(output_path, class_folder)
    os.makedirs(output_class_path, exist_ok=True)
    
    images = glob.glob(os.path.join(class_path, "*"))
    print(f"Processing images for: {class_folder}...")
    
    for img_path in tqdm(images):
        try:
            img = cv2.imread(img_path)
            if img is None: continue
            
            # Convert to grayscale
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            
            # Resize to 256x256
            gray_resized = cv2.resize(gray, (256, 256))
            
            # Convert back to 3-channel grayscale
            gray_3ch = cv2.merge([gray_resized, gray_resized, gray_resized])
            
            # Save to output folder
            base_name = os.path.basename(img_path)
            save_path = os.path.join(output_class_path, base_name)
            cv2.imwrite(save_path, gray_3ch)
        except Exception as e:
            print(f"Error processing {img_path}: {e}")
```

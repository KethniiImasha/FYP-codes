```python
# Training for 7 classes on Apple Silicon
from ultralytics import YOLO
import torch

# 1. Load the pre-trained YOLO11 Nano classification model
model = YOLO('yolo11n-cls.pt') 

# 2. Train the model
# NOTE: device='mps' leverages the M2's GPU/Neural Engine
results = model.train(
    data='/Users/kethniimasha/Documents/Yolo training/AffectNet 6emo data set', 
    epochs=300,                  
    imgsz=256,                   
    batch=64,                    # Lowered from 128; M2 memory management handles 32-64 better for stability
    patience=30,                 
    device='mps',                # Use 'mps' for MacBook M2 GPU acceleration
    workers=4                    # Use multiple CPU cores for data loading
)
```
